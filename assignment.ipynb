Following the table we have take the initial pressure ratio between the driver section and driven section as $25\left(\mathbf{p}_4 / \mathbf{p}_1\right)$.

The strength of the shock which is produced is determined by the pressure ratio of the driver gas to that of the driven, $\mathrm{p} 4 / \mathrm{p} 1$. It also depends on the ratio of sound speeds, $(\mathrm{a} 4 / \mathrm{a} 1$, and the isentropic exponents, $y 4, y 1$, in the two gases. The following equation permits calculation of the strength of the shock produced as a function of the initial conditions:
$$
\frac{p_4}{p_1}=\frac{p_2}{p_1}\left[1-\frac{a_1\left(\gamma_4-1\right)\left(p_2 / p_1-1\right)}{\left.a_4\left(\gamma_1-1\right)\left[\frac{2 \gamma_1}{\gamma_1-1} \frac{p_2}{p_1}\left(\frac{\gamma_1+1}{\gamma_1-1}\right)+\frac{p_1}{p_2}\right]^{1 / 2}\right]-\left(2 \gamma_4\right) /\left(\gamma_4^{-1)}\right.}\right]^{-1}
$$
It may easily be deduced from Equation that stronger shocks are obtained for larger initial pressure ratios across the diaphragm, $\mathrm{p} 2 / \mathrm{p} 1$, for larger values of $\mathrm{a} 4 / \mathrm{a} 1$ and for larger gamma 1. As the speed of sound is inversely proportional to the square root of the molecular weight, it follows that lighter driver gases such as $\mathrm{He}$ and $\mathrm{H}_2$ give much stronger shocks than $\mathrm{N}_2$, say, for the same p4/p1. It is obvious that significantly higher shock strengths are obtained when $\mathrm{H}_2$ or He are used.

It is recalled that gamma is $1.67$ for a monatomic gas, $1.4$ for a diatomic gas with no vibrational heat capacity, and as the molecules become more complex gamma approaches unity. It is seen that the pressure ratio is much less sensitive to gamma, while the temperature and density ratios are extremely sensitive. Regarding the temperature, the physical reasoning leading to the indicated dependence on gamma is as follows: from the normal shock-wave relations it is seen that the specific enthalpy and not the temperature appears explicitly. It can be shown that the specific enthalpy ratio across the shock is nearly independent of gamma for a given shock Mach number, $\mathrm{M}_{\mathrm{w}}$. Temperature is obtained from enthalpy by dividing by $\mathrm{cp}$. For more complex molecules cp is larger since these molecules can absorb more energy in internal degrees of freedom so that the translational temperature $\mathrm{T}_2$ is correspondingly smaller than for a monatomic or diatomic gas. As an example, consider a shock of Mach number of ten which propagates into Argon (gamma $=1.67)$ or ideal air $($ gamma $=1.4)$ at room temperature $\left(\mathrm{T}_1=300^{\circ} \mathrm{K}\right)$. For Argon, the post-shock temperature would be 9000 " $\mathrm{K}$, while for the ideal air it would be approximately $6000^{\circ} \mathrm{K}$. This is one of the reasons that mixtures of a test gas and a monatomic gas (typically Argon) are often used; viz., to increase the test gas temperature without resorting to a stronger shock.

```python
import numpy as np

a4 = 348.95 # initial condition of velocity of sound in region 4
u3 = 399.63
a3 = 269.0015
n = 4 # no. of c- characteristic line in simple wave region
N = int(n*(n+1)/2) # Total number of points in non-simple wave region
xt = np.linspace(-a4, u3-a3, n) # x/t
tx = 1/xt # t/x
```

```python
# velocity of fluid
u = np.zeros(int(N+1)) 
for i in range(1,n+1):
    u[i]= 0.83*(348.95+xt[i-1])

#Calculation of speed of sound in simple region
a= np.zeros(int(N+1)) # Speed of sound at the points in non simple region
    
for i in range(1,n+1):
    a[i] = 348.95 - (0.2*u[i])
    
#J_ invariant calculation    
jm = np.zeros(N+1) # J_minus Riemann invariants
    
for i in range(1,n+1):
    jm[i] = u[i]-(2*a[i])//0.4
    
# Speed of sound in non-simple region
s = 1
x = n
for i in range(1,n):
    s=s+x
    a[s] = (-0.2)*jm[i+1]
    x=x-1
    
# Calculation of J+ Riemann Invariants
jp = np.zeros(N+1) # J_Plus Riemann Invariants
for i in range(1,N+1):
    jp[i] = u[i] + (a[i]/0.2)
    
#########
s = n+2
x=n-2
for i in range(2,n):
    for j in range(s,s+x):
        jp[j]=jp[s-1]
    s=s+x+1
    x=x-1
for i in range(1, n+1):
    x=n-1
    s=i+x
    for j in range(1,i):
        jm[s]=jm[i]
        x=x-1
        s=s+x
        
        
########
for i in range(1,N+1):
    u[i] = 0.5*(jp[i]+jm[i])
    a[i] = 0.1*(jp[i]-jm[i])
    
print ("U value: ",u)
print("a value: ",a)

########

pt = np.zeros((N+1,2))
for i in range(1,n+1):
    if(i==1):
        pt[i][0] = -20
        pt[i][1] = tx[i-1] * pt[i][0]
    else:
        k = (0.5)*((1.0/(u[i]-a[i])) + (1.0/(u[i-1] - a[i-1])) )
        pt[i][0] = (k*pt[i-1][0] - pt[i-1][1])/(k - tx[i-1])
        pt[i][1] = tx[i-1] * pt[i][0]
# print (pt)
```
